```python
def pastel(graf):
    colores=[]
    colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
    for i in colors.keys():
        colores.append(i)
    colores=np.random.choice(colores, 32, replace=False)
    fig1, ax1 = plt.subplots()
    fig1.set_size_inches(18.5, 10.5)
    if 'Inntervalo de clase' in graf.columns:
        ax1.pie(graf['Distribución de porcentajes'], labels=graf['Inntervalo de clase'], colors=colores,autopct='%1.1f%%',
        shadow=True, startangle=90)
    else:
        ax1.pie(graf['Distribuci¢n de porcentajes'], labels=graf['Datos'], colors=colores,autopct='%1.1f%%',
        shadow=True, startangle=90)
    ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    plt.title("Gráfica de pastel")

    plt.show()
```

```python
def histograma(graf):
    if 'Inntervalo de clase' in graf.columns:
        x_pos = [i for i in graf['Inntervalo de clase']]
    else:
        x_pos = [i for i in graf['Datos']]
    plt.figure(figsize=(15,8))
    plt.bar(x_pos,graf['Frecuencia'],width=0.9,color='green')
    plt.title("Histograma")
    if 'Inntervalo de clase' in graf.columns:
        plt.xlabel("Calificaciones finales")
    else:
        plt.xlabel("Tiempo de espera")
    plt.ylabel("Frecuencia")
    plt.show()
```

```python
def barras(graf):
    import numpy as np
    import matplotlib.pyplot as plt

      # the x locations for the groups
    width = 0.27       # the width of the bars

    fig = plt.figure(figsize=(15,8))
    ax = fig.add_subplot(111)

    if 'Inntervalo de clase' in graf.columns:
        x_pos = [i for i in graf['Inntervalo de clase']]
        N = 5
        ind = np.arange(N)
    else:
        N = 21
        ind = np.arange(N)
        x_pos = [i for i in graf['Datos']]

    rects1 = ax.bar(ind, graf['Frecuencia'], width, color='r')
    rects2 = ax.bar(ind+width, graf['Frecuencia acumulada'], width, color='g')

    ax.set_xticks(ind+width)
    ax.set_xticklabels( x_pos )
    ax.legend( (rects1[0], rects2[0]), ('Relativa', 'Acumulada',) )

    def autolabel(rects):
        for rect in rects:
            h = rect.get_height()
            ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                    ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    plt.title("Gráfica de barras")
    if 'Inntervalo de clase' in graf.columns:
        plt.xlabel("Calificaciones finales")
    else:
        plt.xlabel("Tiempo de espera")
    plt.ylabel("Frecuencia relativa y acumulada")

    plt.show()
```

```python
def pol_fec(graf):
    plt.figure(figsize=(15,8))
    if 'Inntervalo de clase' in graf.columns:
        x_pos = [i for i in graf['Inntervalo de clase']]
    else:
        x_pos = [i for i in graf['Datos']]
    plt.plot(x_pos, graf['Frecuencia'], 'ro')
    
    if 'Inntervalo de clase' in graf.columns:
        for i,z in zip(graf['Inntervalo de clase'], graf['Frecuencia']):
            plt.text(x=i,y=z+0.1,s=z, ha='right', va='top')
    else:
        for i,z in zip(datos['Datos'], datos['Frecuencia']):
            plt.text(x=i,y=z+0.1,s=z, ha='right', va='top')

    plt.title("Poligono de frecuencias")
    if 'Inntervalo de clase' in graf.columns:
        plt.xlabel("Calificaciones finales")
    else:
        plt.xlabel("Tiempo de espera")
    plt.ylabel("Frecuencia")

    plt.show()
```

```python
def ojiva(graf):
    plt.figure(figsize=(15,8))

    if 'Inntervalo de clase' in graf.columns:
        plt.plot('Inntervalo de clase', 'Frecuencia acumulada', data=graf, color="red", linestyle='dashed',marker='o', linewidth=2)
        for i,z in zip(graf['Inntervalo de clase'], graf['Frecuencia acumulada']):
            plt.text(x=i,y=z+0.1,s=z, ha='right', va='baseline')  
    else:
        plt.plot( 'Datos', 'Frecuencia acumulada', data=graf, color="red", linestyle='dashed',marker='o', linewidth=2)
        for i,z in zip(graf['Datos'], graf['Frecuencia acumulada']):
            plt.text(x=i,y=z+0.1,s=z, ha='right', va='baseline')


    plt.title("Ojiva")
    if 'Inntervalo de clase' in graf.columns:
        plt.xlabel("Calificaciones finales")
    else:
        plt.xlabel("Tiempo de espera")
    plt.ylabel("Frecuencia acumulada")    

    plt.show()
```
